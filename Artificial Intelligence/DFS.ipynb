```python
def dfs (graph, src, tgt):
    path = []
    stack = []
    stack.append(src)
    while stack:
        node = stack.pop()
        if node not in path:
            path.append(node)
            if node == tgt:
                break
            stack.extend(graph[node])

    return path

test = {
    "a": ["b", "f"],
    "b": ["a", "c", "g"],
    "c": ["b", "d", "g", "l"],
    "d": ["c", "e", "k"],
    "e": ["d", "f"],
    "f": ["a", "e"],
    "g": ["b", "c", "h", "l"],
    "h": ["g", "i"],
    "i": ["h", "j", "k"],
    "j": ["i", "k"],
    "k": ["d", "i", "j", "l"],
    "l": ["c", "g", "k"],
}

dfs(test, "a", "k")
```
